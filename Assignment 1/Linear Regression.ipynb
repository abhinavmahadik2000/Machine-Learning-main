## 2.1 Model Definition

### Simple Linear Regression with mini - batch gradient descent

```python
#Class definition
class LinearRegression():
    def  __init__(self, bias = 0):
        self.bias = bias
        self.weight = None
    
    def fit(self, X, y_actual, epochs = 100, batch_size = 32, lr = 0.01):
        #No. of features will be 1 in training set
        n_features = X.shape[1]
    
        #initialize weight as 1 and bias as 0
        weight = np.ones(shape = (n_features))
        bias = self.bias
    
        #total records
        n_records = X.shape[0]
        
        #lists to record the loss
        mse_list = []
        step_num_list = []
    
        #no. of batches
        n_batches = int(n_records/batch_size)
    
        # In every itteration
        for i in range(epochs):
        
            # Generate array of random values of size equal to n_records
            rand_vals = np.random.permutation(n_records)
        
            # Use these values as indices
            X_tmp = X[rand_vals]
            y_tmp = y_actual[rand_vals]
        
            # to create batches and calculate GD in batches
            for batch in range(0, n_records, batch_size):
                X_batch = X_tmp[batch : batch+batch_size]
                y_batch = y_tmp[batch : batch+batch_size]
            
                # Calculate predictions using the equations for that batch
                # weight * x is done with matrices weight and X uisng dot product
                y_pred = (weight @ X_batch.T) + bias
                # y_pred = np.dot(weight, X_batch.T) + bias   # wx + b
            
                # Calculate new values for weight and bias using derivate equations
                new_weight =  -(2/len(X_batch)) * (X_batch.T @ (y_batch - y_pred))
                #new_weight =  -(2/len(X_batch)) * (X_batch.T.dot(y_batch - y_pred))
                new_bias = -(2/len(X_batch))*np.sum(y_batch-y_pred)
            
                # Update the weight and bias with the new values
                weight = weight - lr * new_weight
                bias = bias - lr * new_bias
            
                # Calculate the MSE
                mse = np.mean((y_batch - y_pred)**2)
        
            # Save the mse and step number
            mse_list.append(mse)
            step_num_list.append(i)        
        
        return weight, bias, mse, mse_list, step_num_list
    
    def predict(self, X, weight, bias):
        predictions = []

        for i in range(len(X)):
            y_pred = weight * X[i] + bias
            predictions.append(y_pred)

        return predictions
    
    def plot_loss(self, mse_list, step_num_list):
        plt.figure(figsize=(12,10))
        plt.plot(step_num_list, mse_list)
        plt.scatter(step_num_list, mse_list, marker="*", c= '#bcbd22', cmap='Greens', edgecolors="Red")
        plt.xlabel("Step number", fontsize = 12)
        plt.ylabel("Loss", fontsize = 12)
        plt.show()
        plt.show()
        
    def accuracy(self, pred, y):
        loss = np.mean((y - pred)**2)
        return loss
```

### Linear Regression with L2 regularization (Ridge Regression)

```python
class LinearRegression_L2():
    def  __init__(self, bias = 0):
        self.bias = bias
        self.weight = None
    
    def fit(self, X, y_actual, L2 = 1, epochs = 100, batch_size = 32, lr = 0.01):
        #No. of features will be 1 in training set
        n_features = X.shape[1]
    
        #initialize weight as 1 and bias as 0
        weight = np.ones(shape = (n_features))
        bias = self.bias
    
        #total records
        n_records = X.shape[0]
        
        #lists to record the loss
        mse_list = []
        step_num_list = []
    
        #no. of batches
        n_batches = int(n_records/batch_size)
    
        # In every itteration
        for i in range(epochs):
        
            # Generate array of random values of size equal to n_records
            rand_vals = np.random.permutation(n_records)
        
            # Use these values as indices
            X_tmp = X[rand_vals]
            y_tmp = y_actual[rand_vals]
        
            # to create batches and calculate GD in batches
            for batch in range(0, n_records, batch_size):
                X_batch = X_tmp[batch : batch+batch_size]
                y_batch = y_tmp[batch : batch+batch_size]
            
                # Calculate predictions using the equations for that batch
                # weight * x is done with matrices weight and X uisng dot product
                y_pred = (weight @ X_batch.T) + bias
                
                # Calculate new values for weight and bias using derivate equations
                new_weight =  -(2/len(X_batch)) * (X_batch.T.dot(y_batch - y_pred)) + ((L2) * (weight**2)) 
                new_bias = -(2/len(X_batch))*np.sum(y_batch-y_pred)
            
                # Update the weight and bias with the new values
                weight = weight - lr * new_weight
                bias = bias - lr * new_bias
            
                # Calculate the MSE
                mse = np.mean((y_batch - y_pred)**2)
        
            # Save the mse and step number
            mse_list.append(mse)
            step_num_list.append(i)        
        
        return weight, bias, mse, mse_list, step_num_list
    
    def predict(self, X, weight, bias):
        predictions = []

        for i in range(len(X)):
            y_pred = weight * X[i] + bias
            predictions.append(y_pred)

        return predictions
    
    def plot_loss(self, mse_list, step_num_list):
        plt.figure(figsize=(12,10))
        plt.plot(step_num_list, mse_list)
        plt.scatter(step_num_list, mse_list, marker="*", c= '#bcbd22', cmap='Greens', edgecolors="Red")
        plt.xlabel("Step number", fontsize = 12)
        plt.ylabel("Loss", fontsize = 12)
        plt.show()
        plt.show()
        
    def accuracy(self, pred, y):
        loss = np.mean((y - pred)**2)
        return loss
```
